```python
# ----------------------------------------------------------
# Population vs Sample Analysis ‚Äî Study Hours Example
# ----------------------------------------------------------

import numpy as np
from scipy import stats

# -------------------------------
# 1Ô∏è‚É£ Define the population & sample
# -------------------------------
population = np.array([8, 10, 6, 12, 9, 14, 7, 11, 13, 10])
sample = np.array([8, 14, 10, 11])

# -------------------------------
# 2Ô∏è‚É£ Population statistics
# -------------------------------
N = len(population)
pop_mean = np.mean(population)
pop_var = np.var(population, ddof=0)       # population variance (divide by N)
pop_std = np.std(population, ddof=0)       # population standard deviation

# -------------------------------
# 3Ô∏è‚É£ Sample statistics
# -------------------------------
n = len(sample)
sample_mean = np.mean(sample)
sample_var = np.var(sample, ddof=1)        # unbiased sample variance (divide by n-1)
sample_std = np.std(sample, ddof=1)        # sample standard deviation
sem = stats.sem(sample, ddof=1)            # standard error of mean

# -------------------------------
# 4Ô∏è‚É£ Display summary
# -------------------------------
print("üìä Population Statistics")
print(f"  N (size): {N}")
print(f"  Mean (Œº): {pop_mean:.2f}")
print(f"  Variance (œÉ¬≤): {pop_var:.2f}")
print(f"  Std Deviation (œÉ): {pop_std:.2f}\n")

print("üìà Sample Statistics")
print(f"  n (size): {n}")
print(f"  Mean (xÃÑ): {sample_mean:.2f}")
print(f"  Variance (s¬≤): {sample_var:.2f}")
print(f"  Std Deviation (s): {sample_std:.2f}")
print(f"  Standard Error (SEM): {sem:.2f}")
```
