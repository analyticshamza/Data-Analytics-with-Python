```python
plt.figure(figsize=(5,5))
plt.plot(df1['amount'])
plt.show()
```

```python
plt.figure(figsize=(5, 5))
plt.bar(df1.index, df1['amount'])
plt.xlabel('Transaction ID (Index)')
plt.ylabel('Amount')
plt.title('Bar Plot of Transaction Amounts')
plt.show()
```

```python
plt.figure(figsize=(5, 3))
plt.plot(df1.index, df1['amount'],'bo-')
plt.xlabel('Range')
plt.ylabel('Amount')
plt.title('Amount Distribution')
plt.show()
```

```python
type_sums = df1.groupby('type')['amount'].sum()
plt.figure(figsize=(5, 5))
plt.pie(type_sums.values, labels=type_sums.index, autopct='%1.1f%%', colors=['#FF9999', '#66B2FF'])
plt.title('Proportion of Amount by Transaction Type')
plt.show()
```

```python
plt.figure(figsize=(5, 3))
plt.plot(df1.index, df1['amount'],color='black', marker='o', linestyle='dashed',linewidth=1.5, markersize=12)
plt.xlabel('Range')
plt.ylabel('Amount')
plt.title('Amount Distribution')
plt.show()
```
